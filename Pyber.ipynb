```python
# 5.1.3 - Line Charts
# Plot a line chart - first define the X and Y variables and values
# Define savings as variable equal to a list
# Define months as a variable set to a list of months as string
savings = [0,100,150,200,300,500,350,500,600,750,1000, 2000]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# No spaces between ' and ,
```

```python
# 5.1.3
# These methods need to be called before the show method in order for chart to display
plt.plot(months,savings)
plt.show()

# Now that we have defined the X and Y arguments, we can pass them in the plot () method
# please note that the order of the arguments matter. The X value will be in the first position and Y value in the second.
# plt.show() shows the current plot
```
